```python
hex_values_per_country = []
no_of_colors_per_flag = [2, 4, 4, 4, 3, 2, 3, 3, 2, 3, 3, 2, 5, 4, 3, 2, 3, 4, 2, 3]

for i, flag in enumerate(flags):
#     response = requests.get('colors/' + flag)
#     img = Image.open(BytesIO(response.content))

    img = Image.open('colors/' + flag)
        
    pixels = img.convert('RGB')

    dict_colors = {}
    for x in range(img.size[0]):
        for y in range(img.size[1]):
            color = pixels.getpixel((x,y))
            if color in dict_colors:
                dict_colors[color] += 1
            else:
                dict_colors[color] = 1
                
    sorted_colors = sorted(dict_colors.items(), key=operator.itemgetter(1))[::-1]
    sorted_colors = list(sorted_colors)
    
    hex_values = []
    for rgb in sorted_colors[:no_of_colors_per_flag[i]]:
#         print(rgb[0])
        hex_values.append('#%02x%02x%02x' % rgb[0])
        
    hex_values_per_country.append(hex_values)

hex_values_per_country
```
